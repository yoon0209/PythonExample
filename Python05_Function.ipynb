```python
def echo_generator():
     while True:
         received = yield
         print("Received:", received)
gen = echo_generator()
next(gen) # Prime the generator
gen.send("Hello")
gen.send("World")
```

```python
def simple_gen():
     yield "Emil"
     yield "Tobias"
     yield "Linus"
gen = simple_gen()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def count_up_to(n):
     count = 1
     while count <= n:
         yield count
         count += 1
for num in count_up_to(5):
     print(num)
```

```python
def fibonacci(n):
     if n <= 1:
         return n
     else:
         return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(7))
```
